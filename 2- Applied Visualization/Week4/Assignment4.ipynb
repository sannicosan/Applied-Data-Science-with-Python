```python
aper_2018=pd.read_csv('aper2018.csv',index_col=0)
aper_2018.index.name='Equipos'
# top 4 team in Paraguay
top_4_names=['Olimpia','Cerro Porteño','Libertad','Guaraní']
aper_2018_top4=aper_2018.loc[top_4_names]
print('\t\t\t\tApertura Tournament 2018')
aper_2018_top4
```

```python
clau_2018=pd.read_csv('clau2018.csv',index_col=0)
clau_2018.index.name='Equipos'
# top 4 team in Paraguay
top_4=['Olimpia','Cerro Porteño','Libertad','Guaraní']
clau_2018_top4=clau_2018.loc[top_4]
print('\t\t\t\tClausura Tournament 2018')
clau_2018_top4
```

```python
aper_2019=pd.read_csv('aper2019.csv',index_col=0)
aper_2019.index.name='Equipos'
# top 4 team in Paraguay
top_4=['Olimpia','Cerro Porteño','Libertad','Guaraní']
aper_2019_top4=aper_2019.loc[top_4]
print('\t\t\t\tApertura Tournament 2019')
aper_2019_top4
```

```python
clau_2019=pd.read_csv('clau2019.csv',index_col=1)
clau_2019.index.name='Equipos'
clau_2019=clau_2019.drop('Equipo\Fecha',axis=1)
# top 4 team in Paraguay
top_4=['Olimpia','Cerro Porteño','Libertad','Guaraní']
clau_2019_top4=clau_2019.loc[top_4]
print('\t\t\t\tClausura Tournament 2019')
clau_2019_top4
```

```python
top4_2018 = pd.merge(aper_2018_top4,clau_2018,how='left',left_index=True,right_index=True)
top4_2019 = pd.merge(aper_2019_top4,clau_2019,how='left',left_index=True,right_index=True)
top4_df = pd.merge(top4_2018,top4_2019,how='left',left_index=True,right_index=True)
top4_df.columns = list(np.arange(1,top4_df.shape[1]+1))
top4_df = top4_df.transpose()
tour_init_idx=[1,23,45,67]
top4_df.loc[tour_init_idx]=np.nan
top4_df.index.name='Matches'
#top4_df = top4_df.fillna(method='bfill')
top4_df.head()
```

```python
top4_df=top4_df.reset_index()
```

```python
# Figure initialization
fig, ((ax1,ax2) , (ax3,ax4)) = plt.subplots(2,2,sharex=True,sharey=True,figsize=(11,6))
fig.suptitle('Top4 Soccer Teams Standings [2018-2019]',fontsize=14)
fig.subplots_adjust(left=0.07,wspace=0.1,bottom=0.2)
axes=[ax1,ax2,ax3,ax4]
colors=['k','b','g','y']
tour_init_idx.append(89)

# Drawing data
for j,ax in enumerate(axes):
    # Main Club plot
    ax.plot(top4_df['Matches'],top4_df[top_4_names[j]],color=colors[j],linewidth=2.5)
    ax.scatter(x=tour_init_idx[1:],y=(top4_df.iloc[np.array(tour_init_idx[1:])-2,j+1].values),color=colors[j],marker='*',s=75)
    ax.vlines(x=tour_init_idx[1:],ymin=1,ymax=12,color=colors[j],alpha=0.2)
    ax.set_title(top_4_names[j])
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    for i in range(1,4):
        # Other club in background
        ax.plot(top4_df['Matches'],top4_df[top_4_names[i]],color='gray',alpha=0.1)
        ax.scatter(x=tour_init_idx[1:],y=(top4_df.iloc[np.array(tour_init_idx[1:])-2,i+1].values),color='gray',alpha=0.4,marker='*')
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        

# Axes customization
ax.invert_yaxis()
ax.set_yticks(np.arange(1,13));
ax.set_yticklabels([str(i)+'º' for i in range(1,14)])
ax.set_xticks(np.arange(1,92,4));

fig.text(0.45,0.12,'Matches',fontsize=12);
fig.text(0.01,0.6,'Standings',rotation='vertical',fontsize=12) ;
fig.text(s='*Data points correspond to standings in the period 2018-2019 of soccer teams ' \
            'from Asunción-Paraguay (1 season = half year = 22 matches).\n'\
            '* The v-lines show the beggining of a new season\n'\
            '* The stars show the standings at the end of the season',\
             x=0.08,y=0.025, color='gray',alpha=0.5, fontsize=8 );
```
