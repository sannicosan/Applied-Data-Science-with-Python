In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection as LC 
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
%matplotlib notebook

# Parameters from data (mean,sem, confidence interval)
years=list(df.index)

# Years Standard Errors of the Mean
year_sem = df.sem(axis=1).values
year_sem = ['{:0.1f}'.format(y) for y in year_sem]
year_sem = np.array(year_sem,dtype='float32')
# Mutiply by 1.96 to get confidence interval:
year_conf = 1.96*year_sem

# Years Means
year_mean = df.mean(axis=1).values
year_mean = list('{:0.1f}'.format(y) for y in year_mean)
year_mean = np.array(year_mean,dtype='float32')
```

```python
# To Draw the bars

def draw_bars(fig_idx):
    
    global fig,ax,bars
    
    # Figure initialization
    fig=plt.figure(fig_idx,figsize=(10,7))
    
    ax = fig.add_subplot(1,1,1)
    ax.set_title('Custom Sampled Distribution',y=1.05,fontsize=14)

    ax.set_xticks(years)
    ax.set_xlabel('Years',fontsize=12,labelpad=1)

    ax.set_yticks(np.arange(0,51000,2500))
    ax.set_ylabel('Value',fontsize=12,labelpad=10)

    # Bars
    bars=ax.bar(years,year_mean,yerr=year_conf,color='lightskyblue',capsize=10)

    # Customization
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    ax.set_yticklabels(np.arange(0,51000,2500),color='lightskyblue',picker=20)
    ax.tick_params(axis='y',color='lightskyblue')
    ax.set_ylim(top=50000,emit=False)
    
    #figure size
    figsize=fig.get_size_inches()*fig.dpi
    ax.annotate('Select Value\n\t'+r'$\Downarrow$',xy=[0.035*figsize[0],0.65*figsize[1]],xycoords='figure points')
```

```python
# For Event : Picking y-axis value for bar coloring

#----------------------------------------------------
# Remove previous h-line picked (= y-value selected)
def remove_hline(axes):
    if isinstance(axes.get_children()[1],LC):
        axes.get_children()[1].remove() 
        
#------------------------------------------------------
# colormapping depending on y-value selected
def bar_compare(y_selected):
    #global which_list
    
    # Colormap to color bars
     # Creating custom colormap (BWR Colormap w/ alpha=0.8)
    alpha=0.9
    n_colors=128
    cmap=cm.get_cmap(name='bwr',lut=n_colors)
    newcolors=cmap(np.linspace(0,1,n_colors))
    newcolors[:,3]=alpha
    my_cmap=ListedColormap(newcolors)
    
    # Constructing all confidence intervals
    conf_inter = list( zip(year_mean-year_conf,year_mean+year_conf) )
    
    # Calculating color value from my_cmap
    color_tol=0.1
    which_list= []
    for k in range(len(conf_inter)):
        
        if( y_selected > conf_inter[k][1]):
            y_range = np.abs( ax.get_ylim()[1] - conf_inter[k][1] )
            y_level_selected = np.abs( y_selected - conf_inter[k][1])
            color_range = color_tol+0.5
            bar_color = my_cmap( int( ( color_range + (y_level_selected/y_range)*(1-color_range) ) *n_colors ) )
            which='encima'
        # Por debajo
        elif( y_selected < conf_inter[k][0]):
            y_range = np.abs( ax.get_ylim()[0] - conf_inter[k][0] )
            y_level_selected = np.abs( y_selected - ax.get_ylim()[0])
            color_range = 0.5-color_tol
            bar_color = my_cmap( int( ((y_level_selected/y_range)*(color_range) )*n_colors ) )
            which='debajo'

        # Dentro del intervalo de confianza
        else:
            y_range = np.abs( conf_inter[k][1] - conf_inter[k][0])
            y_level_selected = np.abs( y_selected - conf_inter[k][0] )
            color_range = 0.5-color_tol
            bar_color = my_cmap(int( (color_range + (y_level_selected/y_range)*(2*color_tol) )*n_colors ))
            which='dentro'
        
        #which_list.append(which)
        bars[k].set_color(bar_color)
        
#------------------------------------------------------  
# Event implementation
def on_pick(event):
    
    global y_selected
    y_selected=event.artist.get_position()[1]
    
    ax.set_title('Selected Level=' + str(y_selected))
    ax_xlims=ax.get_xlim() 
   
    # Removing previous lines
    remove_hline(ax)
    
    ax.hlines( int(y_selected), ax.get_xlim()[0] ,ax.get_xlim()[1] ,alpha=0.5)
    ax.set_xlim(left=ax_xlims[0],right=ax_xlims[1])
    
    bar_compare(y_selected)  
```

```python
# Run to draw bars based on data
draw_bars(1)
```

```python
# Run to Select Y-Value from Y-axis and color the Bars consequently [Even Harder Option]
picked = plt.gcf().canvas.mpl_connect('pick_event', on_pick)
```
